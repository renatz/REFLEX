```python
import json
def read_json_object(file_path):
    with open(file_path, 'r', encoding="utf-8") as file:
        json_data = json.load(file)
    return json_data
```

```python
split = 'dev'
# mode = 'cevi'
mode = 'cev'
data = read_json_object(f"/home/comp/24483737/fc/data/post-averitec/last/dev_averitec_cev_ordinal.json")
# data = read_json_object(f"/home/comp/24483737/fc/data/post-averitec/last/dev_averitec_cevi_ordinal.json")
```

```python

remove = read_json_object(f"/home/comp/24483737/fc/data/post-averitec/remove/dev_averitec_cevi_cross.json")
```

```python
remove_idx = []
for sample in remove:
    remove_idx.append(sample['id'])
```

```python
remain_list = []
for sample in data:
    if sample['id'] in remove_idx:
        continue
    remain_list.append(sample)
```

```python
with open(f'./post-averitec/lalast/{split}_averitec_{mode}_ordinal_remove.json', 'w', encoding="utf-8" ) as file:
    json.dump(remain_list, file, indent=2, ensure_ascii=False)
```
