```python
def read_single_line(file_path):
    with open(file_path, 'r', encoding="utf-8") as file:
        session_list = file.readlines()
    new_list = []
    for session_str in session_list:
        new_list.append(eval(session_str))
    return new_list
```

```python
base = read_single_line(f"{INPUT_DIR}{BASE_INPUT_FILE}")
sft = read_single_line(f"{INPUT_DIR}{SFT_INPUT_FILE}")
```

```python
import numpy as np
def compute_length(data:list):
    len_l=[]
    for sample in data:
        a = sample['choices'][0]['turns'][0] 
        len_l.append(len(a))
    # print(max(len_l))
    print('Mean length', np.mean(len_l))
```

```python
compute_length(base)
```

```python
compute_length(sft)
```

```python
def fix_unicode_escape(text):
    return text.encode("latin1", errors="ignore").decode("utf-8", errors="ignore")
```

```python
def clean(data):
    cleaned = []
    null_cnt = 0
    for sample in data:
        a = sample['choices'][0]['turns'][0] 
        fixed = fix_unicode_escape(a)
        fixed = fixed.replace('\n',' ')
        sample['choices'][0]['turns'][0] = fixed
        if not sample['choices'][0]['turns'][0]:
            continue
        cleaned.append(sample)
    print(f"Null_counter:{null_cnt}")
    print(f"Final length:{len(cleaned)}")
    return cleaned
```

```python
cleaned_sft = clean(sft)
compute_length(sft)
```

```python
cleaned_base = clean(base)
compute_length(base)
```
