```python
import re
def clean_verdict(text: str) -> str:
    pattern = r"^(Verdict:\s*(?:Supported|Refuted|Not Enough Evidence))(?:\s+\w+)*\."
    return re.sub(pattern, r"\1.", text, flags=re.MULTILINE)
examples = [
    "Verdict: Not Enough Evidence industry. Explanation: The Q&A pairs do not show enough evidence to support or refute the claim.",
    "Verdict: Supported industries. Explanation: The claim is supported.",
    "Verdict: Refuted blah blah. Explanation: Not true.",
    "Verdict: Not Enough Evidence. Explanation: Already clean."
]

for e in examples:
    print("Before:", e)
    print("After: ", clean_verdict(e))
    print()
```

```python
def make_merged_pairs(data):
    sv_sample = []
    for sample in data:
        sample_dict = {}
        sample_dict['question'] = sample['question']
        ctgr = sample['category']
        if ctgr == 'hallucinated':
            if 'mistral' in INPUT_PATH:
                if 'averitec' in INPUT_PATH:
                    pattern = r"^(Verdict:\s*(?:Supported|Refuted|Not Enough Evidence))(?:\s+\w+)*\."
                elif 'liaraw' in INPUT_PATH:
                    pattern = r"^(Verdict:\s*(?:TRUE|FALSE|HALF-TRUE))(?:\s+\w+)*\."
                else:
                    pattern = r"^(Verdict:\s*(?:true|false|half))(?:\s+\w+)*\."
                    
                cleaned_sft = re.sub(pattern, r"\1.", sample['sft_answer'], flags=re.MULTILINE)
                sample_dict['answer_not_matching_behavior'] = cleaned_sft
                cleaned_base = re.sub(pattern, r"\1.", sample['base_answer'], flags=re.MULTILINE)
                sample_dict['answer_matching_behavior'] = cleaned_base
            else:
                sample_dict['answer_not_matching_behavior'] = sample['sft_answer']
                sample_dict['answer_matching_behavior'] = sample['base_answer']

        elif ctgr == 'success': 

            sample_dict['answer_not_matching_behavior'] = sample['base_answer']
            if not sample['sft_answer'].startswith('Verdict: '):
                res = re.sub(r".*?(Verdict.*)", r"\1", sample['sft_answer'])
                sample_dict['answer_matching_behavior'] = res
            else:
                sample_dict['answer_matching_behavior'] = sample['sft_answer']

            if 'mistral' in INPUT_PATH:
                if 'averitec' in INPUT_PATH:
                    pattern = r"^(Verdict:\s*(?:Supported|Refuted|Not Enough Evidence))(?:\s+\w+)*\."
                elif 'liaraw' in INPUT_PATH:
                    pattern = r"^(Verdict:\s*(?:TRUE|FALSE|HALF-TRUE))(?:\s+\w+)*\."
                else:
                    pattern = r"^(Verdict:\s*(?:true|false|half))(?:\s+\w+)*\."
                cleaned_sft = re.sub(pattern, r"\1.", sample['sft_answer'], flags=re.MULTILINE)
                sample_dict['answer_matching_behavior'] = cleaned_sft
                cleaned_base = re.sub(pattern, r"\1.", sample['base_answer'], flags=re.MULTILINE)
                sample_dict['answer_not_matching_behavior']  = cleaned_base
            else:
                sample_dict['answer_matching_behavior'] = sample['sft_answer']
                sample_dict['answer_not_matching_behavior'] = sample['base_answer']

        sv_sample.append(sample_dict)
    return sv_sample
```
